```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x ,y1, label='Person 1')
ax.plot(x ,y2, label='Person 2')
ax.legend(loc = 'lower right')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x ,y1, label='Person 1')
ax.plot(x ,y2, label='Person 2')
ax.set_title("people availability over time")
ax.set_xlabel('time')
ax.set_ylabel('People')
ax.legend(loc = 'lower right')
fig.show()

```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x ,y1, label='Person 1')
ax.plot(x ,y2, label='Person 2')
ax.legend(loc = 'lower right', frameon = True, #add frame around the legend
        fontsize = 12, #change font size
        ncol = 2, #specify number of columns
        shadow = True )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x ,y1, label='Person 1')
ax.plot(x ,y2, label='Person 2')
ax.legend(loc = 'lower right', bbox_to_anchor =(1,1))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x ,y1, label='Person 1')
ax.scatter(x ,y2, label='Person 2')
ax.legend(loc = 'lower right', bbox_to_anchor =(1,1))
ax.text(10, 95, 'This value is important!', ha='center', fontsize=20, color='red',
        bbox=dict(facecolor='yellow', alpha=0.5))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x ,y1, label='Person 1')
ax.scatter(x ,y2, label='Person 2')
ax.legend(loc = 'lower right', bbox_to_anchor =(1,1), prop={"family":"serif"})
ax.text(10, 95, 'This value is important!', ha='center', fontsize=20, color='red', fontfamily='serif')
        
fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)",   transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.1, 0.2)", transform=fig.transFigure)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
          arrowprops=dict(facecolor='black'))
fig.show()

```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), arrowprops = dict(arrowstyle = "wedge",color = "blue"))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) 
ax.yaxis.set_major_locator(plt.MaxNLocator(5)) 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(10)) 
```
